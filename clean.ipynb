# A Group with Exactly one noncommutator

## Commutator Calculator Function

```gap
numComm := function(G)
    # Check if g belongs to the derived subgroup using the theorem
    local numCommutator, chi, tbl, sm, i, classes; 
    numCommutator := 0;
    tbl := CharacterTable(G);
    classes := ConjugacyClasses(G);


    #Print("Table: ", Irr(tbl), "\n");
    for i in [1..NrConjugacyClasses(G)] do
        #Print("New Class: ", i, "\n\n");
        sm := 0;
        for chi in Irr(tbl) do
        #Print("Character = ", chi, "\n");
            #Print(chi[i], " / ", chi[1], "\n");
            sm := sm + (chi[i] / chi[1]);
        od;
        #Print("sum = ", sm,  "\n");
        if sm = 0 then
            # Print(classes[i], "\n");
            numCommutator := numCommutator + Size(classes[i]);
        fi;
    od;

    
    
    return numCommutator;
end;
```

## Search for Group Action $\varphi$

### Find $N \cong \mathbb{Z}_2^{10}$

```gap
normals := NormalSubgroups(pg1);
List(normals, Order);

N := normals[10];
Elements(N);
```

### Find $H \cong A_6$

```gap
# get A6
A6 := AlternatingGroup(6);

#Get aut(N)
autN := AutomorphismGroup(N);
```

Get a subgroup isomorphic to $A_6$ and check $H \cap N = \emptyset$

```gap
H := Image(IsomorphicSubgroups(pg1, other)[1]);
Intersection(H, N);
```

### Create $\varphi ' : H \to \operatorname{Aut}(N)$

```gap
#generators of H 
gens_H := GeneratorsOfGroup(H);
#generators of N 
gens_N := GeneratorsOfGroup(N);
#image of each generator is conjugation 
imgs := List([1..Size(gens_H)], i -> GroupHomomorphismByImages(N,N,gens_N,List([1..Size(gens_N)], j -> gens_N[j]^gens_H[i])));

hom := GroupHomomorphismByImages(H,autN,gens_H,imgs);
```

Test that $N \rtimes H$ has one noncommutator

```gap
test := SemidirectProduct(H, hom, N);
```

```gap
n := Order(test);
chi := numComm(test);
Print("The value of the length for group ", n, " is: ", CommutatorLength(test)," and its center is of order ", Order(Center(test)),"\n");
Print("The number of noncommutators is ", chi, " and the order of the group is ", n,"\n");

StructureDescription(test:nice);
```

### Create ismorphism $\theta : \operatorname{Aut}(N) \to \operatorname{Aut}(\mathbb{Z}_2^{10})$

Create group $\mathbb{Z}_2^{10}$

```gap
Z2 := CyclicGroup(2);

Z210 := DirectProduct(C2, C2, C2, C2, C2, C2, C2, C2, C2, C2);
Random(Z210);
```

It is too large to use `IsomorphismGroups` immediately between the groups $\operatorname{Aut}(N)$ and $\operatorname{Aut}(\mathbb{Z}_2^{10})$, so we will compute an isomorphism using an isomoprhism $\theta' : N \to \mathbb{Z}_2^{10}$.

```gap
aut_iso := IsomorphismGroups(autN, autZ210);
```

```gap
theta_prime := IsomorphismGroups(N, Z210);
```

We create a custom `function` to calculate the compositon

```gap
theta := GroupHomomorphismByFunction(autN,autZ210, 
    function(f) return CompositionMapping(theta_prime, f, InverseGeneralMapping(theta_prime)); end);
```
